```python
def clean_html(page):
    
    
    review = []
    
    soup_desc = BeautifulSoup(page['desc_html'], "html.parser")
    
    try:
        
        title = soup_desc.find('div',attrs = {'class':'bloc-title'}).text
        description = soup_desc.find('div',attrs = {'class':'descriptif'}).text
    
    except:
        return
    
    
    
    try:
    
        soup_review = BeautifulSoup(page['reviews_html'], "html.parser")
        
        block = soup_review.find('div',attrs = {'class':'bloc-reviews'})
        
        reviews = block.find_all('div',attrs = {'class':'d-none'})
        
        for x in reviews:            
            review.append(x.text.strip())
            
    except Exception as e:
        print(e)
        review = None
    
    return {'title':title,'description':description,'reviews':review}
    
```

```python
max_workers = 10

with ThreadPoolExecutor(max_workers=max_workers) as executor:
    futures = [executor.submit(clean_html, x) for x in data]
    
    for future in tqdm(futures, total=len(data)):
        result = future.result()
        if result:
            with open('data/tryandreview-parsed.jsonl', 'a') as f:
                json.dump(result, f)
                f.write('\n')
```
